### pandas 모듈을 이용하여 파일의 내용을 DataFrame으로 생성한다.

```python
import pandas as pd # 통상적으로 pd라고 씀
from pandas import DataFrame
```

```python
# txt 문서를 읽어 DataFrame 구하기 // 한글 깨지면 해당 파일 UTF-8로 다시 저장하기
#              읽어올 파일명                 구분 기호
dataFrame = pd.read_table('testFile/test_text_file.txt', '|', engine='python')
display(dataFrame)
dataFrame
```

```python
dataFrame2 = pd.read_table('testFile/naverHotKey.txt', ',', engine='python')
display(dataFrame2) # 너무 길면 중간 내용 생략됨
display(dataFrame2.head(10)) # 상단 10개까지만 담김
```

```python
# title이 파일에 없는 경우 // header 이름은 List 형식으로 담아준다.
dataFrame3 = pd.read_table('testFile/test_text_file2.txt','|', header=None, names=['이름','국어','영어','수학','과학'], engine='python')
display(dataFrame3)
```

```python
# dataFrame에서 컬럼명 얻어오기
columnList = dataFrame3.columns
columnList
```

```python
# 컬럼 삭제 
data1 = dataFrame3.drop('과학', axis=1) # 행 지정: 0, 열 지정: 1
display(data1)
```
